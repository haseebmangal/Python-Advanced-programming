# fibonic sries

```python
def sam1(n):
    x=0
    y=1
    print(x,y)
    for i in range(2,n):
        z=x+y
        print(z)
        x=y
        y=z 
```

```python
sam1(8)
```

```python
def feb1(n):
    if(n<=1):
        return n
    
    return feb1(n-1) +  feb1(n-2)
```

# lcm and hcf

```python
def lcm1(x,y):
    if(x>y):
        greater=x
    else:
        greater=y
    while(1):
        if(greater%x==0 and greater%y==0):
            break
        greater=greater+1
    return greater
    
```

```python
lcm1(6,9)
```

```python
def lcm(x,y):
    if(x>y):
        greater=x
    else:
        greater=y
    while(1):
        if(greater%x==0 and greater%y==0):
            break
        greater=greater+1
    return greater    
```

```python
lcm(6,9)
```

```python
def hcf(x,y):
    if(x<y):
        smaller=x
    else:
        smaller=y
    while(1):
        if(x%smaller==0 and y%smaller==0):
            break
        smaller-=1
    return smaller    
            
```

```python
hcf(6,9)
```

```python
def sample(a):
    le=len(a)-1
    b=[]
    while(le>=0):
        b.append(a[le])
        le=le-1
    return b    
           
```

```python
a=[3,2,2,5,4,3,2]
sample(a)
```
