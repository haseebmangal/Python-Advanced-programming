```python
def roundup(x):
    remainder=x%1
    if remainder < 0.5:
        print("number rounded down")
        x=x-remainder
    else:
        print("number is rounded up")
        x=x+(1-remainder)
    return x
        
```

```python
roundup(364.4)
```

```python
def prime2(x):
    result=0
    for y in range(2,x):
        if x%y==0:
            result=1
            print(x,"number is not prime number")
            break
    if(result==0):
        print(x,"number is prime")
```
